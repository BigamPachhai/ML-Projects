```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(["deck",'embark_town','alive','class','who','adult_male'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.dropna(subset=['embarked'],inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.head()
```

```python
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
df=df.astype(int)
```

```python
df.head()
```

```python
x=df.drop('survived',axis=1)
y=df['survived']
```

# **Logistic Regression**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.preprocessing import  StandardScaler
scaler=StandardScaler()
```

```python
X_tarined_scaled= scaler.fit_transform(x_train)
X_test_scaled= scaler.transform(x_test)
```

# **KNN-K Nearest Neighbour**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_tarined_scaled,y_train)
```

```python
y_pred_knn=knn_model.predict(X_test_scaled)
```

```python
accuracy_score(y_test,y_pred_knn)
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
print(classification_report(y_test,y_pred_knn))
```

# **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_NB= GaussianNB()
model_NB.fit(x_train,y_train)
```

```python
y_pred_NB=model_NB.predict(x_test)
```

```python
y_pred_NB
```

```python
accuracy_score(y_test,y_pred_NB)
```

```python
confusion_matrix(y_test,y_pred_NB)
```

```python
print(classification_report(y_test,y_pred_NB))
```

# **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DT=DecisionTreeClassifier(random_state=45)
model_DT.fit(X_tarined_scaled,y_train)
```

```python
y_pred_DT=model_DT.predict(X_test_scaled)
```

```python
y_pred_DT
```

```python
accuracy_score(y_test,y_pred_DT)
```

```python
confusion_matrix(y_test,y_pred_DT)
```

```python
print(classification_report(y_test,y_pred_DT))
```

# **SVM- Support Vector Machine**

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(kernel='linear')
model_svm.fit(X_tarined_scaled,y_train)
```

```python
y_pred_svc=model_svm.predict(X_test_scaled)
```

```python
accuracy_score(y_test,y_pred_svc)
```

```python
confusion_matrix(y_test,y_pred_svc)
```

```python
print(classification_report(y_test,y_pred_svc))
```
